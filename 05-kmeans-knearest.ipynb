```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans

X = np.array([[1, 6], [2, 7], [3, 6], [4, 2], [5, 3],
            [6, 1], [7, 5], [8, 6], [9, 4], [10, 5]])

#plt.scatter(X[:,0],X[:,1], label='True Position')

kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
print(kmeans.cluster_centers_)
plt.scatter(X[:,0],X[:,1], c=kmeans.labels_, cmap='rainbow')

#plt.xlim(0, 9)
#plt.ylim(0, 6.5)
plt.scatter(kmeans.cluster_centers_[:,0],
            kmeans.cluster_centers_[:,1], color='black')

plt.show()
```

```python
from sklearn.neighbors import NearestCentroid
import numpy as np
X = np.array([[3, 2], [5, 3], [2, 1],
              [2, 2], [6, 3], [6, 4]])
y = np.array(['N', 'P', 'N',
              'N', 'P', 'P'])
plt.scatter(X[:,0],X[:,1])
clf = NearestCentroid()
clf.fit(X, y)

print(clf.predict([[4, 4]]))
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from sklearn.cluster import KMeans

X = np.array([[10,8,10],[10,0,9],[4,8.5,3],[9.5,0.5,8.5],[4.5,8.5,3],
              [10.5,9,12],[5,8.5,11],[2.7,8.7,2],[9.7,2,9],[10.2,1,9.2]])

kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
print(kmeans.cluster_centers_)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0],X[:,1],X[:,2], c=kmeans.labels_, cmap='rainbow', marker='o')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans

X = np.array([[1, 2], [1, 3], [4, 5], [5, 5]])

#plt.scatter(X[:,0],X[:,1], label='True Position')

kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
print(kmeans.cluster_centers_)
plt.scatter(X[:,0],X[:,1], c=kmeans.labels_, cmap='rainbow')

plt.xlim(0, 6)
plt.ylim(0, 6)
plt.scatter(kmeans.cluster_centers_[:,0],
           kmeans.cluster_centers_[:,1], color='black')

plt.show()
```
