# Ejemplo de Scikit-learn para reconocimiento de images

Se realizará un entrenamiento de Support Vector Classifier SVC utilizando el dataset de digits, luego se entrenará dejando el último dato de prueba para luego predecirlo.

```python
# importing the hand written digit dataset
from sklearn import datasets

# digit contain the dataset
digits = datasets.load_digits()

# dir function use to display the attributes of the dataset
dir(digits)
```

```python
import matplotlib.pyplot as plt

_, axes = plt.subplots(nrows=1, ncols=10, figsize=(4, 3))
for ax, image, label in zip(axes, digits.images, digits.target):
    ax.set_axis_off()
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation="nearest")
    ax.set_title("%i" % label)
```

```python
digits.target
```

```python
from sklearn import svm
clf = svm.SVC(gamma=0.001, C=100.)
clf.fit(digits.data[:-1], digits.target[:-1])
```

```python
clf.predict(digits.data[-1:])
```
